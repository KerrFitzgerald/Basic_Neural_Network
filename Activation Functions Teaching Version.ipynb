# Implementing Activation functions

# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define sigmoid function

```python
def sigmoid(Z):
    a = 1.0/(1.0 + np.exp(-Z))
    return a
```

# Test sigmoid function

```python
X = np.arange(-10.0,10.0, 0.1, dtype='float')
Y = sigmoid(X)
fig = plt.figure()
plt.plot(X,Y)
plt.show()
```

# Define sigmoid derivative function

```python
def sig_deriv(Z):
    a = sigmoid(Z)*(1.0 - sigmoid(Z))
    return a
```

# Test sigmoid derivative function

```python
X = np.arange(-10.0,10.0, 0.1, dtype='float')
Y = sig_deriv(X)
fig = plt.figure()
plt.plot(X,Y)
plt.show()
```

# Define tanh activation function

```python
def tanh(Z):
    a = np.tanh(Z)
    return a
```

# Test tanh activation function

```python
X = np.arange(-10.0,10.0, 0.1, dtype='float')
Y = tanh(X)
fig = plt.figure()
plt.plot(X,Y)
plt.show()
```

# Define tanh derivative function

```python
def tanh_deriv(Z):
    a = 1.0 - np.power(np.tanh(Z),2)
    return a
```

# Test tanh activation function

```python
X = np.arange(-10.0,10.0, 0.1, dtype='float')
Y = tanh_deriv(X)
fig = plt.figure()
plt.plot(X,Y)
plt.show()
```

# Define softmax function

```python
def softmax(Z):
    a = np.exp(Z - np.max(Z))
    b = a/a.sum(axis=0, keepdims=True)
    return b
```
